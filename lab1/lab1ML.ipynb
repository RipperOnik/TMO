## 1) Текстовое описание набора данных

В качестве набора данных мы будем использовать набор данных Wine из Scikit-learn.
Этот датасет является результатом химического анализа вин, выращенных в одном регионе Италии тремя разными культиваторами. Было проведено тринадцать различных измерений для различных составляющих трех типов вина. 
Содержатся следующие колонки: 

- Alcohol

- Malic acid

- Ash

- Alcalinity of ash

- Magnesium

- Total phenols

- Flavanoids

- Nonflavanoid phenols

- Proanthocyanins

- Color intensity

- Hue

- OD280/OD315 of diluted wines

- Proline



## Импорт библиотек

```python
import numpy as np
import pandas as pd
from sklearn.datasets import *
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

## Загрузка данных 

```python
wine = load_wine()
```

```python
type(wine)
```

```python
for x in wine:
    print(x)
```

```python
wine['feature_names']
```

```python
data1 = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])

```

```python
# Проверим наличие пустых значений
# Цикл по колонкам датасета
for col in data1.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data1[data1[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

## 3) Визуальное исследование датасета

### Диаграмма рассеяния 

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='total_phenols', y='flavanoids', data=data1)
```

Как видно из графика, проследуется линейная зависимость между 2 компонентами.

### Гистограмма

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data1['alcohol'])
```

### Jointplot 

```python
sns.jointplot(x='flavanoids', y='total_phenols', data=data1)

```

```python
sns.jointplot(x='flavanoids', y='total_phenols', data=data1, kind="hex")
```

## 4) Информация о корреляции признаков

С целевым признаком коррелируют nonflavanoid_phenois, malic_acid, alcalinity_of_ash. Также существует отрицательная корреляция с total_phenois, flavanolds, hue, proline. Это значит, что рост целевой компоненты приводит к уменьшению других компонентов.
